```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
x = np.concatenate(x,axis=0)
y = np.concatenate(y,axis=0)
```

```python
df = pd.DataFrame({
                    'x': x,
                    'y': y
                })
df
```

```python
def pop_initialization(p_size):
    p = []
    for i in range(p_size):
        rand_arr = np.random.uniform(-5,5, size=4)
        p.append(rand_arr)
    return p
```

```python
def fitness_score(x,y,p):
    fitness_arr = np.zeros(len(p))
    y_pred = np.zeros(len(y))
    for p_i in range(0,len(p)):
        for x_i in x:
            for y_i in range(0,len(y)):
                y_pred[y_i] =  p[p_i][0] * x_i**3 + p[p_i][1] * x_i**2 + p[p_i][2] * x_i + p[p_i][3]
                break
        fitness_arr[p_i] = np.negative(np.sum(y-y_pred)**2)
    return fitness_arr
```

```python
p = pop_initialization(5)
print(p)
fitness_arr = fitness_score(x,y,p)
print(fitness_arr)

ind = np.argsort(fitness_arr)[-2:]
print(ind)

temp_new_arr_ind = 0
num_new_arrs = 2
all_new_arrs = []
print('p:', p)
for i in ind:
    for j in range(0,num_new_arrs):
        temp_new_arr = []
        print('new_arr:', temp_new_arr)
        print('i:',i)
        temp_new_arr = p[i]
        print('new_arr',temp_new_arr)
        rand_ind = np.random.randint(len(temp_new_arr))
        print('rand_ind:',rand_ind)
        rand_float = np.random.uniform(0,2)
        print('rand_float:',rand_float)
        temp_new_arr[rand_ind] = temp_new_arr[rand_ind] * rand_float
        print('modified new_arr:',temp_new_arr)
        print('p[i]:', p[i])
        all_new_arrs.append(temp_new_arr)
    print(all_new_arrs)        
```
